```python
# %% [markdown]
# # Windows Event Log Exploration & Initial Analysis
# 
# This notebook performs initial exploration of Windows Event Logs to understand:
# - Event type distributions
# - Temporal patterns
# - Session characteristics
# - Common sequences
# - Data quality issues

# %%
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime, timedelta
import re
import os
import xml.etree.ElementTree as ET
from collections import Counter
import plotly.express as px

# Setup
%matplotlib inline
sns.set(style="whitegrid", palette="pastel")
plt.rcParams['figure.figsize'] = (12, 6)
pd.set_option('display.max_columns', 50)
pd.set_option('display.max_colwidth', 100)

# %% [markdown]
# ## 1. Data Loading and Parsing

# %%
# Define column mapping for malware dataset
column_map = {
    'SystemTime': 'TimeCreated',
    'EventID': 'Id',
    'Level': 'LevelDisplayName',
    'ProviderName': 'ProviderName',
    'Description': 'Message',
    'Channel': 'LogName'
}

# Define level mapping for malware dataset
level_name_map = {
    0: 'Information',
    1: 'Critical',
    2: 'Error',
    3: 'Warning',
    4: 'Information',
    5: 'Verbose'
}

# %%
def load_normal_logs(file_path):
    """Load and preprocess normal logs"""
    print(f"Loading normal logs from {file_path}")
    # Read with low_memory=False to handle mixed types
    df = pd.read_csv(file_path, low_memory=False)
    
    # Clean and preprocess
    df = df.dropna()
    df['TimeCreated'] = pd.to_datetime(df['TimeCreated'], format='%d-%m-%Y %H:%M:%S', errors='coerce')
    df = df.dropna(subset=['TimeCreated'])
    
    # Standardize column names
    df = df.rename(columns={
        'TimeCreated': 'timestamp',
        'Id': 'event_id',
        'LevelDisplayName': 'level',
        'ProviderName': 'source',
        'Message': 'message',
        'LogName': 'log_name'
    })
    
    return df

def load_malware_logs(file_path):
    """Load and preprocess malware logs"""
    print(f"Loading malware logs from {file_path}")
    # Read with low_memory=False to handle mixed types
    df = pd.read_csv(file_path, low_memory=False)
    
    # Rename columns
    df = df.rename(columns=column_map)
    
    # Handle duplicate columns (common in large CSV files)
    df = df.loc[:, ~df.columns.duplicated()]
    
    # Convert levels
    if 'LevelDisplayName' in df.columns:
        df['LevelDisplayName'] = pd.to_numeric(df['LevelDisplayName'], errors='coerce')
        df['LevelDisplayName'] = df['LevelDisplayName'].map(level_name_map).fillna('Unknown')
    
    # Convert timestamps
    if 'TimeCreated' in df.columns:
        df['TimeCreated'] = pd.to_datetime(df['TimeCreated'], errors='coerce')
        df = df.dropna(subset=['TimeCreated'])
    
    # Convert event IDs - handle as series
    if 'Id' in df.columns:
        # Ensure we're working with a Series
        id_series = df['Id']
        df['Id'] = pd.to_numeric(id_series, errors='coerce').astype('Int64')
    
    # Fill empty messages
    if 'Message' in df.columns:
        df['Message'] = df['Message'].fillna('No message available')
    
    # Standardize column names
    column_mapping = {
        'TimeCreated': 'timestamp',
        'Id': 'event_id',
        'LevelDisplayName': 'level',
        'ProviderName': 'source',
        'Message': 'message',
        'LogName': 'log_name'
    }
    
    # Only map columns that exist in the DataFrame
    final_mapping = {k: v for k, v in column_mapping.items() if k in df.columns}
    df = df.rename(columns=final_mapping)
    
    # Create missing columns if needed
    for col in ['timestamp', 'event_id', 'level', 'source', 'message', 'log_name']:
        if col not in df.columns:
            df[col] = None
    
    return df

# %%
# Load datasets
normal_logs = load_normal_logs('../data/raw/evtx_logs/WindowsEventLogs_Last90Days.csv')
malware_logs = load_malware_logs('../data/raw/evtx_logs/malware_data.csv')

# Add labels
normal_logs['label'] = 'normal'
malware_logs['label'] = 'malware'

# Combine for analysis
combined_logs = pd.concat([normal_logs, malware_logs], ignore_index=True)

print("Dataset sizes:")
print(f"Normal logs: {len(normal_logs):,} events")
print(f"Malware logs: {len(malware_logs):,} events")
print(f"Combined: {len(combined_logs):,} events")

# Show first few rows of each
print("\nNormal logs sample:")
display(normal_logs[['timestamp', 'event_id', 'level', 'source']].head(2))

print("\nMalware logs sample:")
display(malware_logs[['timestamp', 'event_id', 'level', 'source']].head(2))

# %% [markdown]
# ## 2. Data Quality Check

# %%
def data_quality_report(df):
    """Generate data quality report"""
    report = pd.DataFrame({
        'dtype': df.dtypes,
        'missing_values': df.isnull().sum(),
        'missing_pct': df.isnull().mean() * 100,
        'unique_values': df.nunique()
    })
    
    # Add examples for message field
    if 'message' in df.columns:
        report.loc['message', 'examples'] = " | ".join(df['message'].sample(3, random_state=42).tolist())
    
    return report

# %%
print("Normal Logs Data Quality:")
display(data_quality_report(normal_logs))

print("\nMalware Logs Data Quality:")
display(data_quality_report(malware_logs))

# %% [markdown]
# ## 3. Event Type Distribution Analysis

# %%
# Event ID distribution
def plot_event_distribution(df, title):
    plt.figure(figsize=(14, 8))
    event_counts = df['event_id'].value_counts().head(20)
    ax = sns.barplot(x=event_counts.values, y=event_counts.index, palette="viridis")
    plt.title(f"Top 20 Event IDs: {title}")
    plt.xlabel("Count")
    plt.ylabel("Event ID")
    
    # Add percentages
    total = len(df)
    for i, v in enumerate(event_counts.values):
        ax.text(v + total*0.01, i, f"{v/total:.1%}", ha='left', va='center')
    
    plt.tight_layout()
    plt.savefig(f"../reports/figures/event_distribution_{title.lower().replace(' ', '_')}.png")

# %%
plot_event_distribution(normal_logs, "Normal Activity")
plot_event_distribution(malware_logs, "Malware Activity")

# %% [markdown]
# ### Event Level Comparison

# %%
level_comparison = pd.crosstab(combined_logs['label'], combined_logs['level'], normalize='index') * 100

plt.figure(figsize=(10, 6))
sns.heatmap(level_comparison, annot=True, fmt=".1f", cmap="YlGnBu")
plt.title("Event Level Distribution (%)")
plt.ylabel("Dataset")
plt.xlabel("Level")
plt.tight_layout()
plt.savefig("../reports/figures/event_level_comparison.png")

# %% [markdown]
# ## 4. Temporal Pattern Analysis

# %%
def plot_temporal_patterns(df, title):
    # Resample to hourly - updated to 'h' from 'H'
    df = df.copy()
    df.set_index('timestamp', inplace=True)
    hourly = df.resample('h').size()
    
    # Daily pattern
    daily = df.groupby(df.index.hour).size()
    
    # Weekly pattern
    df['dayofweek'] = df.index.dayofweek
    weekly = df.groupby('dayofweek').size()
    
    fig, ax = plt.subplots(3, 1, figsize=(14, 12))
    
    # Hourly plot
    hourly.plot(ax=ax[0], title=f"Hourly Event Count: {title}", color='teal')
    ax[0].set_xlabel("")
    ax[0].set_ylabel("Count")
    
    # Daily pattern
    daily.plot(ax=ax[1], kind='bar', title=f"Daily Pattern: {title}", color='skyblue')
    ax[1].set_xlabel("Hour of Day")
    ax[1].set_ylabel("Average Count")
    
    # Weekly pattern
    weekly.plot(ax=ax[2], kind='bar', title=f"Weekly Pattern: {title}", color='salmon')
    ax[2].set_xlabel("Day of Week")
    ax[2].set_ylabel("Total Count")
    ax[2].set_xticklabels(['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun'])
    
    plt.tight_layout()
    plt.savefig(f"../reports/figures/temporal_patterns_{title.lower().replace(' ', '_')}.png")

# %%
plot_temporal_patterns(normal_logs, "Normal Activity")
plot_temporal_patterns(malware_logs, "Malware Activity")

# %% [markdown]
# ## 5. Session Analysis

# %%
def analyze_sessions(df, max_gap_minutes=5):
    """Analyze session characteristics from logs"""
    print(f"Analyzing sessions for {len(df):,} events...")
    
    # Sort by timestamp
    df = df.sort_values('timestamp').copy()
    
    # Calculate time differences
    df['time_diff'] = df['timestamp'].diff().dt.total_seconds() / 60
    
    # Identify new sessions
    df['new_session'] = (df['time_diff'] > max_gap_minutes) | df['time_diff'].isna()
    df['session_id'] = df['new_session'].cumsum()
    
    # Session analysis
    sessions = df.groupby('session_id').agg(
        start_time=('timestamp', 'min'),
        end_time=('timestamp', 'max'),
        duration_min=('timestamp', lambda x: (x.max() - x.min()).total_seconds() / 60),
        event_count=('event_id', 'count'),
        unique_events=('event_id', 'nunique'),
        error_count=('level', lambda x: (x == 'Error').sum())
    )
    
    return sessions.reset_index()

# %%
# Analyze sessions
normal_sessions = analyze_sessions(normal_logs)
malware_sessions = analyze_sessions(malware_logs)

print("\nNormal Session Statistics:")
display(normal_sessions.describe())

print("\nMalware Session Statistics:")
display(malware_sessions.describe())

# %% [markdown]
# ### Session Duration Comparison

# %%
plt.figure(figsize=(10, 6))
sns.kdeplot(normal_sessions['duration_min'], label='Normal', fill=True)
sns.kdeplot(malware_sessions['duration_min'], label='Malware', fill=True)
plt.title("Session Duration Distribution")
plt.xlabel("Duration (minutes)")
plt.xlim(0, 60)  # Focus on first 60 minutes
plt.legend()
plt.savefig("../reports/figures/session_duration_comparison.png")

# %% [markdown]
# ### Event Count per Session

# %%
plt.figure(figsize=(10, 6))
sns.boxplot(x='label', y='event_count', 
            data=pd.concat([
                normal_sessions.assign(label='normal'),
                malware_sessions.assign(label='malware')
            ]))
plt.title("Event Count per Session")
plt.yscale('log')  # Log scale for better visualization
plt.savefig("../reports/figures/session_event_count.png")

# %% [markdown]
# ## 6. Common Event Sequences

# %%
def find_common_sequences(df, n=3, top_k=10):
    """Find most common event sequences of length n"""
    sequences = []
    df = df.sort_values('timestamp')
    
    # Group by session or use entire log if no sessions
    group_key = 'session_id' if 'session_id' in df.columns else None
    
    if group_key:
        for _, group in df.groupby(group_key):
            events = group['event_id'].tolist()
            for i in range(len(events) - n + 1):
                sequences.append(tuple(events[i:i+n]))
    else:
        events = df['event_id'].tolist()
        for i in range(len(events) - n + 1):
            sequences.append(tuple(events[i:i+n]))
    
    # Count frequencies
    counter = Counter(sequences)
    return counter.most_common(top_k)

# %%
# Find common sequences in normal logs
print("Top 10 Normal Sequences (3 events):")
normal_sequences = find_common_sequences(normal_logs)
for seq, count in normal_sequences:
    print(f"{' → '.join(map(str, seq))}: {count}")

# Find common sequences in malware logs
print("\nTop 10 Malware Sequences (3 events):")
malware_sequences = find_common_sequences(malware_logs)
for seq, count in malware_sequences:
    print(f"{' → '.join(map(str, seq))}: {count}")

# %% [markdown]
# ### Sequence Visualization

# %%
def plot_sequence_network(sequences, title, max_nodes=20):
    """Plot event sequence network with size constraints"""
    import networkx as nx
    
    G = nx.DiGraph()
    
    # Add edges with weights
    for seq, count in sequences[:max_nodes]:
        for i in range(len(seq) - 1):
            source = seq[i]
            target = seq[i+1]
            if G.has_edge(source, target):
                G[source][target]['weight'] += count
            else:
                G.add_edge(source, target, weight=count)
    
    # Draw graph with constrained size
    plt.figure(figsize=(10, 7))
    pos = nx.spring_layout(G, k=0.5)
    
    # Node size by degree
    node_size = [d * 100 for n, d in G.degree()]
    
    # Edge width by weight
    edge_width = [d['weight'] / 100 for u, v, d in G.edges(data=True)]
    
    nx.draw_networkx_nodes(G, pos, node_size=node_size, node_color='lightblue', alpha=0.9)
    nx.draw_networkx_edges(G, pos, width=edge_width, edge_color='gray', alpha=0.7)
    nx.draw_networkx_labels(G, pos, font_size=8)
    
    plt.title(f"Event Sequence Network: {title}")
    plt.axis('off')
    plt.tight_layout()
    plt.savefig(f"../reports/figures/sequence_network_{title.lower().replace(' ', '_')}.png")

# %%
# Plot with size constraints
plot_sequence_network(normal_sequences, "Normal Activity")
plot_sequence_network(malware_sequences, "Malware Activity")

# %% [markdown]
# ## 7. Text Message Analysis

# %%
from wordcloud import WordCloud

def generate_message_wordcloud(df, title):
    # Combine all messages
    text = " ".join(df['message'].astype(str))
    
    # Generate word cloud
    wordcloud = WordCloud(width=800, height=400, 
                          background_color='white',
                          stopwords=['the', 'and', 'to', 'of', 'a', 'in', 'for', 'on', 'with']
                         ).generate(text)
    
    # Plot
    plt.figure(figsize=(14, 8))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.title(f"Common Message Terms: {title}")
    plt.savefig(f"../reports/figures/wordcloud_{title.lower().replace(' ', '_')}.png")

# %%
generate_message_wordcloud(normal_logs, "Normal Activity")
generate_message_wordcloud(malware_logs, "Malware Activity")

# %% [markdown]
# ### Malware-specific Keywords

# %%
def detect_malware_keywords(df):
    """Identify potential malware-related keywords in messages"""
    malware_keywords = [
        'powershell', 'script', 'commandline', 'rundll32', 'wmi', 
        'registry', 'scheduled', 'uacme', 'proxy', 'dll', 'remote',
        'sysmon', 'whoami', 'cryptographic', 'injection', 'payload',
        'beacon', 'c2', 'lateral', 'brute'
    ]
    
    results = {}
    for keyword in malware_keywords:
        count = df['message'].str.contains(keyword, case=False).sum()
        results[keyword] = count
    
    return pd.Series(results)

# %%
# Compare keyword frequencies
normal_keywords = detect_malware_keywords(normal_logs)
malware_keywords = detect_malware_keywords(malware_logs)

keyword_df = pd.DataFrame({
    'normal': normal_keywords,
    'malware': malware_keywords
})

# Filter out keywords not present in normal logs
keyword_df = keyword_df[keyword_df['normal'] > 0]
keyword_df['ratio'] = keyword_df['malware'] / keyword_df['normal']

# Get top 15 discriminative keywords
top_kw = keyword_df.sort_values('ratio', ascending=False).head(15)

# Plot
plt.figure(figsize=(12, 6))
sns.barplot(x='ratio', y=top_kw.index, data=top_kw)
plt.title("Top Discriminative Keywords")
plt.xlabel("Frequency Ratio (Malware/Normal)")
plt.ylabel("Keyword")
plt.tight_layout()
plt.savefig("../reports/figures/keyword_ratio_comparison.png")

# %% [markdown]
# ## 8. Initial Insights for Anomaly Detection

# %%
# Summary of key differences
insights = pd.DataFrame({
    'Characteristic': [
        'Error Event Frequency',
        'Average Session Duration',
        'Events per Session',
        'PowerShell Keywords',
        'Network Keywords',
        'Common Sequence 4624→4624',
        'Common Sequence 4663→4663'
    ],
    'Normal': [
        (normal_logs['level'] == 'Error').mean(),
        normal_sessions['duration_min'].mean(),
        normal_sessions['event_count'].mean(),
        normal_keywords.get('powershell', 0) / len(normal_logs),
        normal_keywords.get('remote', 0) / len(normal_logs),
        sum(1 for seq, _ in normal_sequences if seq == (4624, 4624)),
        sum(1 for seq, _ in normal_sequences if seq == (4663, 4663))
    ],
    'Malware': [
        (malware_logs['level'] == 'Error').mean(),
        malware_sessions['duration_min'].mean(),
        malware_sessions['event_count'].mean(),
        malware_keywords.get('powershell', 0) / len(malware_logs),
        malware_keywords.get('remote', 0) / len(malware_logs),
        sum(1 for seq, _ in malware_sequences if seq == (4624, 4624)),
        sum(1 for seq, _ in malware_sequences if seq == (4663, 4663))
    ]
})

insights['Ratio'] = insights['Malware'] / insights['Normal'].replace(0, 0.001)

print("Key Differences Summary:")
display(insights)

# %% [markdown]
# ## 9. Next Steps for Feature Engineering
# 
# Based on this exploration, we'll focus on these features:
# 
# 1. **Markov Transition Probabilities**:
#    - Especially error persistence (Error → Error)
#    - Warning escalation (Warning → Error)
#    
# 2. **Entropy Measures**:
#    - Session-level event diversity
#    - Sequence unpredictability
#    
# 3. **Temporal Patterns**:
#    - Beaconing detection using FFT
#    - Error burst detection
#    
# 4. **Sequence Features**:
#    - Frequency of suspicious sequences (e.g., 4663→4663)
#    - Long non-repeating chains
#    
# 5. **Keyword Features**:
#    - Density of suspicious keywords
#    - Rare command occurrences
#    
# 6. **Session Characteristics**:
#    - Duration
#    - Error rate
#    - Event diversity
# 
# These will be implemented in `02_feature_engineering_EDA.ipynb`.

# %%
# Save processed data for next notebook
normal_logs.to_csv('../data/processed/normal_processed.csv', index=False)
malware_logs.to_csv('../data/processed/malware_processed.csv', index=False)
print("Processed data saved to ../data/processed/")
```
